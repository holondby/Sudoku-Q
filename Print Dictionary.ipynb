Print_dictionary() displays a nested dictionary. This function can generate an extensive amount of screen output, so it should be used sparingly. It has been included primarily as a research tool. 

```python
import os
import numpy as np

def initialize_memory(filename):
    if os.path.exists(filename): 
        memory = np.load(filename,allow_pickle='TRUE').item()
        return memory 
    return {}

def print_dictionary(memory):
    for key,value in memory.items():
        if isinstance(value,dict):
            items = dict(sorted(value.items()))
            actions = list(items.keys())
            q_values = list(items.values())
            best = np.argmax(q_values)
            print("\nState:\n")
            print(key)
            print("\nRecommended action:",actions[best],"\n")              
            print_dictionary(items)
        else:
            formatted_value = "{:+.12f}".format(value)
            print("action:",key,"\tq-value:",formatted_value)
    return

def main(): 
    memory_file = "test_ltm.npy"
    memory = initialize_memory(memory_file)
    print("Memory Contents")
    print_dictionary(memory)
    return

if __name__ == "__main__":
    main()
```
